```python
def plot_gauss(data, space = 'x', name = "gauss"):
    """
    :param data: gaussian data np.array
    :param space: phasespace coordintate, momentum(mom)/position(pos)
    :return:
    """
    plt.rc('text', usetex=False)
    plt.rc('font', family='serif')
    fig = plt.figure(facecolor='w', figsize=(16,9))
    ax = fig.add_subplot(111)
    
    nodes = len(data)
    
    xlabel = "x"
    if space == 'p':
        data = sc.fft(data)/np.sqrt(nodes)
        xlabel = "p"
    
    data_mod2 = np.abs(data)**2
    x0 = (np.argmax(data_mod2))/nodes * 2*pi
    
    ax.plot(np.linspace(0, 2*pi, nodes), data_mod2)
    ax.plot([x0, x0], [0, 1], linewidth = 5)
    
    plt.xticks(np.linspace(0, 2*pi, 5, endpoint=True), ['0', r'$\pi$/2', r'$\pi$', r'3/2$\pi$', r'2$\pi$'],fontsize = 18)
    plt.yticks(fontsize=14)
    ax.set_xlabel(xlabel, fontsize=25)
    ax.set_ylabel(r'$|\psi|^2$', fontsize=25, rotation=0, position=[0.0,0.6])
    
    ax.set_xlim(0, 2*pi)
    ax.set_ylim(0, np.max(data_mod2))
    plt.grid()
    plt.savefig(name+".png")

```

```python
def init_gauss(x0, p0, nodes = 100, num_l = 8):
    '''
    :param x0: initial position of particle [0,2pi)
    :param p0: initial momentum of particle [0,2pi)
    :param nodes: number of nodes in space
    :param Num_l: number of mirror reflections (default = 8)
    :return: normalized gaussian package
    '''

    n = np.arange(nodes)
    n0 = round(nodes/2/pi*x0)
    m0 = round(nodes/2/pi*p0)

    num_l = np.arange(-round(num_l/2), round(num_l/2)+1)
    sums = 0

    for l0 in num_l:
        sums += np.exp(-pi/nodes * (n-n0 + nodes*l0)**2)

    psi = np.exp(1j* 2 * pi * m0 / nodes * n) * sums

    norm = np.linalg.norm(psi)

    return psi / norm
```

```python
gauss = init_gauss(pi/2, 13*pi/7)
plot_gauss(gauss,  space='p', name='gauss_x')
plot_gauss(gauss,  space='x', name='gauss_p')
```

```python
def step(psi0, K = 0):
    '''
    :param psi0: Initial wave function np.array()
    :param K: Driven force coupling
    :return: Wave function after one step
    '''
    nodes = len(psi0)
    n = np.arange(nodes)
    P = np.exp(1j * (pi*n/nodes) * n)
    V = np.exp(-1j * nodes * K / 2/pi *
               np.cos(2 * pi * n/nodes))
    F = sc.fft(P * V * psi0) / np.sqrt(nodes)
    return (P*F)
```

```python
def plot_gauss_axis(data, axis, space = 'x', color = 'r'):
    """
    :param data: gaussian data np.array
    :param space: phasespace coordintate, momentum(mom)/position(pos)
    :return:
    """
    plt.rc('text', usetex=False)
    plt.rc('font', family='serif')
    
    nodes = len(data)
    
    xlabel = "x"
    if space == 'p':
        data = sc.fft(data)/np.sqrt(nodes)
        xlabel = "p"
    
    data_mod2 = np.abs(data)**2
    x0 = (np.argmax(data_mod2))/float(nodes) * 2*pi
    
    axis.set_xticks([])
    axis.set_yticks([])
    axis.plot(np.linspace(0, 2*pi, nodes), data_mod2, color=color)
    #axis.plot([x0, x0], [0, 1], linewidth = 1)
    
    axis.set_xlim(0, 2*pi)
    axis.set_ylim(0, np.max(data_mod2))
    plt.grid()
    return axis
```

```python
times = np.arange(50)

Kn = 7
Nn = 5
x0 = pi/2
dx0 = pi/18
p0 = 3/2*pi
dp0 = 0
f, axarr = plt.subplots(Kn, Nn)
K = np.logspace(-Kn+1, 0, base=2, num=Kn)
Klog = np.log2(K)
N = np.logspace(1, Nn, base=10, num=Nn)
```

```python
psi0 = []
psi1 = []
for k in enumerate(K):
    for n in N:
        psi0.append(init_gauss(x0, p0, nodes = n, num_l = 8))
        psi1.append(init_gauss(x0 + dx0, p0 + dp0, nodes = n, num_l = 8))
```

```python
dx = 0.8/(float(Nn))
dy = 0.8/(float(Kn))
space0 = 'p'

for tn,t in enumerate(times):
    iterr = 0
    f, axarr = plt.subplots(Kn, Nn)
    for kn, k in enumerate(K):
        for nn, n in enumerate(N):
            f.text(0.035 + dx*(nn+1), 0.05, r'$10^'+str(nn+1)+'$', ha='center', fontsize=15)
            f.text(0.075, 0.04 + dy*(kn+1), r'$2^{'+str(int(Klog[kn]))+'}$', ha='center', fontsize=15)
            axarr[Kn-1-kn,nn] = plot_gauss_axis(psi0[iterr], axarr[Kn-1-kn,nn], space = space0, color='b')
            axarr[Kn-1-kn,nn] = plot_gauss_axis(psi1[iterr], axarr[Kn-1-kn,nn], space = space0, color='g')
            psi0[iterr] = step(psi0[iterr],K=k)
            psi1[iterr]= step(psi1[iterr],K=k)
            iterr +=1
    f.text(0.1, 0.94, r'$x_0='+str(round(x0,2))+'\quad p_0 = '+str(round(p0,2))+'\quad dx_0 = '+str(round(dx0,2))+'\quad dp_0 = '+str(round(dp0,2))+'\quad t= '+str(t)+r'\tau_0$'+'   space: '+str(space0))  
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
    f.text(0.5, 0.02, 'Number of nodes M', ha='center')
    f.text(0.03, 0.5, 'K parameter', va='center', rotation='vertical')
    plt.savefig('pltsubplotermfast'+str(tn).zfill(3)+'.png')
```

## Phasespace

```python
paths = 150
xp = np.random.random([paths,2])*2*pi
times = np.arange(70)


K = [2**-4, 2**-2, 2**-0, 2**1, 2**2]
Klog = np.log2(K)
N = [1e1, 1e2, 1e3, 1e4]
f, axarr = plt.subplots(len(K), len(N))
```

```python
dx = 1/(float(len(K)))
dy = 0.8/(float(len(N)+1))
space0 = 'x'
psi0 = []

Kn = len(K)
Nn = len(K)
f, axarr = plt.subplots(len(K), len(N))
for xp0 in xp:
    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
    plt.subplots_adjust(wspace=0, hspace=0)
    
    c = np.random.random([1,3])
    for kn,k in enumerate(K):
        for nn,n in enumerate(N):
            points = []
            psi0 = init_gauss(xp0[0], xp0[1], nodes = n, num_l = 8)
            for tn,t in enumerate(times):
                points.append((np.argmax(psi0)/n*2*pi,np.argmax(sc.fft(psi0))/n*2*pi))
                psi0 = step(psi0, K=k)
            points = np.array(points)
            axarr[Kn-1-kn,nn].scatter(points[:,0],points[:,1], color=c, lw = 0.4)
            axarr[Kn-1-kn,nn].set_xticks([])
            axarr[Kn-1-kn,nn].set_yticks([])
            f.text(0.035 + dx*(nn+1), 0.05, r'$10^'+str(int(np.log10(n)))+'$', ha='center', fontsize=15)
            f.text(0.075, 0.02 + dy*(kn+1), r'$2^{'+str(int(Klog[kn]))+'}$', ha='center', fontsize=15)

    f.text(0.5, 0.02, 'Number of nodes M', ha='center')
    f.text(0.1, 0.94, 'Phase space')
    f.text(0.03, 0.5, 'K parameter', va='center', rotation='vertical')
plt.savefig('paths')
```
